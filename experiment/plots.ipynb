```python
import itertools
import matplotlib.pyplot as plt
```

```python
lines = []
with open("output.txt") as inp:
    for line in inp:
        lines.append(line[:-1])
```

```python
times = { "noopt": {}, "sexeq": {}, "egal": {}, "minreg": {}, "maxcar": {}}
for i in range(int(len(lines) / 7)):
    name = lines[7 * i].split(' ')[2][6:-10]
    
    times["noopt"][name] = float(lines[7 * i + 1].split(' ')[-2])
    times["sexeq"][name] = float(lines[7 * i + 2].split(' ')[-2])
    times["egal"][name] = float(lines[7 * i + 3].split(' ')[-2])
    times["minreg"][name] = float(lines[7 * i + 4].split(' ')[-2])
    times["maxcar"][name] = float(lines[7 * i + 5].split(' ')[-2])
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-0pc-w-0pc".format(m, w)
    vals.append(times["noopt"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- No Optimization (Ties: 0%, Completeness: 100%)")

plt.subplot(2, 3, 2)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-0pc-w-0pc".format(m, w)
    vals.append(times["minreg"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- Minimum Regret (Ties: 0%, Completeness: 100%)")

plt.subplot(2, 3, 3)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-0pc-w-0pc".format(m, w)
    vals.append(times["maxcar"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- Maximum Cardinality (Ties: 0%, Completeness: 100%)")

plt.subplot(2, 3, 4)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-10pc-w-10pc".format(m, w)
    vals.append(times["noopt"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- No Optimization (Ties: 10%, Completeness: 100%)")

plt.subplot(2, 3, 5)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-10pc-w-10pc".format(m, w)
    vals.append(times["minreg"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- Minimum Regret (Ties: 10%, Completeness: 100%)")

plt.subplot(2, 3, 6)
vals = []
for m, w in itertools.product([10, 20, 30], [10, 20, 30]):
    name = "m-{}-w-{}--m-100pc-w-100pc--m-10pc-w-10pc".format(m, w)
    vals.append(times["maxcar"][name])
plt.xticks(range(9), ["10&10", "10&20", "10&30", "20&10", "20&20", "20&30", "30&10", "30&20", "30&30"])
plt.scatter(range(9), vals)
plt.title("Size -- Maximum Cardinality (Ties: 10%, Completeness: 100%)")

plt.show()
```
