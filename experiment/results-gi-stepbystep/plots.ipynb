```python
times = { "noopt": {}, "sexeq": {}, "egal": {}, "minreg": {}, "maxcar": {}}

with open("output-original.txt") as inp:
    for line in inp:
        if line.find("result") != -1:
            name = line.split(' ')[2][6 : line.find("result") - 9]
            
            if line.find("sexequal") != -1: times["sexeq"][name] = float(line.split(' ')[-2])
            elif line.find("egalitarian") != -1: times["egal"][name] = float(line.split(' ')[-2])  
            elif line.find("minregret") != -1: times["minreg"][name] = float(line.split(' ')[-2])
            elif line.find("maxcardinality") != -1: times["maxcar"][name] = float(line.split(' ')[-2])
            else: times["noopt"][name] = float(line.split(' ')[-2])
                
with open("output-gte30.txt") as inp:
    for line in inp:
        if line.find("result") != -1:
            name = line.split(' ')[2][6 : line.find("result") - 9]
            
            if line.find("sexequal") != -1: times["sexeq"][name] = float(line.split(' ')[-2])
            elif line.find("egalitarian") != -1: times["egal"][name] = float(line.split(' ')[-2])  
            elif line.find("minregret") != -1: times["minreg"][name] = float(line.split(' ')[-2])
            elif line.find("maxcardinality") != -1: times["maxcar"][name] = float(line.split(' ')[-2])
            else: times["noopt"][name] = float(line.split(' ')[-2])
            
with open("output-eq120.txt") as inp:
    for line in inp:
        if line.find("result") != -1:
            name = line.split(' ')[2][6 : line.find("result") - 9]
            
            if line.find("sexequal") != -1: times["sexeq"][name] = float(line.split(' ')[-2])
            elif line.find("egalitarian") != -1: times["egal"][name] = float(line.split(' ')[-2])  
            elif line.find("minregret") != -1: times["minreg"][name] = float(line.split(' ')[-2])
            elif line.find("maxcardinality") != -1: times["maxcar"][name] = float(line.split(' ')[-2])
            else: times["noopt"][name] = float(line.split(' ')[-2])
```
